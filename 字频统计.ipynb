```python
import pandas as pd
import re
import plotly.express as px
```

```python
with open ('C:\\Users\\86139\\Desktop\\内科.txt') as f:
    s = f.read()
```

```python
s1 = re.findall(r'[\u4e00-\u9fff]',s)
frequency = pd.Series(list(s1)).value_counts()
frequency1 = frequency.reset_index()
frequency1.columns = ['字符', '频率']
frequency1
```

```python
fig = px.bar(frequency1, x='字符', y='频率', title='字符频率直方图',
             labels={'频率': '出现次数', '字符': '字符'},
             width=1200, height=600)
fig.update_layout(
    xaxis_tickangle=45,  # x轴标签旋转45度
    yaxis_range=[0, max(frequency1['频率']) * 1.1]  # y轴范围设为最高频率的110%
)
fig.show()
```
